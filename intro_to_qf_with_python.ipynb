#Train & Test Data

```python
train_data = df.iloc[0:int(len(data)*0.8), :]
test_data = df.iloc[int(len(data)*0.8):, :]

fig = go.Figure()

fig.add_trace(go.Scatter(x=train_data.index, y=train_data['Open'], mode='lines', name='Training Data', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=test_data.index, y=test_data['Open'], mode='lines', name='Testing Data', line=dict(color='green')))

fig.update_layout(
    title='AAPL Open Prices: Training and Testing Data',
    xaxis_title='Dates',
    yaxis_title='Open Prices',
    template='plotly_white'
)

fig.show()
```

```python
window = 7
train_series = train_data['Open']

#Determing rolling statistics
rolmean = train_series.rolling(window).mean()
rolstd = train_series.rolling(window).std()

#Plot rolling statistics:
fig = plt.figure(figsize=(10, 6))
orig = plt.plot(train_series, color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation');
```

Setting up an [Augmented Dickey-Fuller test](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test) to dtermine if the data is stationary

```python
from statsmodels.tsa.stattools import adfuller

window = 7
train_series = train_data['Close']

dftest = adfuller(train_series, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
dfoutput
```

you can see that the p-value is close to 1, so we fail to reject null hypothesis that the data is not stationary

The stationarity of time series data is crucial because most of the models for forecasting assume that the underlying data are stationary

```python
# calculates the difference between consecutive elements and dropping any na values

train_diff = train_series.diff(periods=1)
train_diff = train_diff.dropna(inplace = False)
```

```python
# determine rolling statistics
window = 7

rolling_mean = train_diff.rolling(window).mean()
rolling_std = train_diff.rolling(window).std()

# plot rolling statistics:
fig = plt.figure(figsize=(10, 6))
orig = plt.plot(train_diff, color='blue',label='Original')
mean = plt.plot(rolling_mean, color='red', label='New Rolling Mean')
std = plt.plot(rolling_std, color='black', label = 'New Rolling Std')
plt.legend(loc='best')
plt.title('New Rolling Mean & Standard Deviation');
```

retrying the [Augmented Dickey-Fuller test](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test) to determine if the data is stationary

```python
dftest = adfuller(train_diff, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
dfoutput
```

Now you can see that the p-value of 1.18 ^ -14 is less than our statistically significant threshold of 0.05 (In other words, the p-value is less than one-tenth of one ten-millionth (0.0000000001)).  We can now reject the null hypothesis that the data is not stationary and proceed.  

```python
# Symmetric Mean Absolute Percentage Error (SMAPE)

def smape_kun(y_true, y_pred, epsilon=1e-8):
    denominator = (np.abs(y_pred) + np.abs(y_true)) + epsilon
    absolute_percentage_error = np.abs(y_pred - y_true) * 200 / denominator
    mean_smape = np.mean(absolute_percentage_error)
    return mean_smape
```

```python
test_series = test_data['Close']
test_diff = test_series.diff(periods=1)
test_diff = test_diff.dropna(inplace = False)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Initialize history with training data
history = [x for x in train_diff]
predictions = list()

# Iterate through the test data points
for t in range(len(test_diff)):
    p, d, q = 5,1,0
    model = ARIMA(history, order=(p, d, q))
    model_fit = model.fit()

    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)

    obs = test_diff[t]
    history.append(obs)

    if t % 100 == 0:
        print('Test Series Point: {}\tPredicted={}, Expected={}'.format(t, yhat, obs))

# Calculate Mean Squared Error (MSE) to evaluate model performance
mse = mean_squared_error(test_diff, predictions)
print('Mean Squared Error (MSE):', mse)
```

calculating cumulative sum

```python
reverse_test_diff = np.r_[test_series.iloc[0], test_diff].cumsum()
reverse_predictions = np.r_[test_series.iloc[0], predictions].cumsum()
reverse_test_diff.shape, reverse_predictions.shape
```

calculating error

```python
error = mean_squared_error(reverse_test_diff, reverse_predictions)
print('Testing Mean Squared Error: %.3f' % error)
error2 = smape_kun(reverse_test_diff, reverse_predictions)
print('Symmetric Mean absolute percentage error: %.3f' % error2)
```

takes the reverse_test_diff & reverse_predictions, and converts them into pandas Series, setting their indices to match that of an existing time series

```python
reverse_test_diff_series = pd.Series(reverse_test_diff)
reverse_test_diff_series.index = test_series.index

reverse_predictions_series = pd.Series(reverse_test_diff)
reverse_predictions_series.index = test_series.index
```

```python
# plotting

plt.figure(figsize=(12,7))
plt.title('IBM Prices')
plt.xlabel('Dates')
plt.ylabel('Prices')
plt.plot(train_series, color='blue', label='Training Prices')
plt.plot(reverse_test_diff_series, color='green', marker='.', label='Testing Prices - Reverse Diff Transform')
plt.plot(reverse_test_diff_series, color='red', linestyle='--', label='Forecasted Prices - Reverse Diff Transform')
plt.legend();
```

moving average crossover:

```python


# genearting aapl data
data = yf.download('AAPL', start='2020-01-01', end='2024-01-01')
data['Fast_MA'] = data['Close'].rolling(window=10).mean()
data['Slow_MA'] = data['Close'].rolling(window=50).mean()

# generate signals
data['Signal'] = 0.0
data['Signal'] = np.where(data['Fast_MA'] > data['Slow_MA'], 1.0, -1.0)
data['Position'] = data['Signal'].diff()

# plot
fig = go.Figure()

# add Close price trace
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='AAPL Close', line=dict(color='blue')))

# add Fast Moving Average trace
fig.add_trace(go.Scatter(x=data.index, y=data['Fast_MA'], mode='lines', name='10-Day MA', line=dict(color='red')))

# add Slow Moving Average trace
fig.add_trace(go.Scatter(x=data.index, y=data['Slow_MA'], mode='lines', name='50-Day MA', line=dict(color='green')))

# add buy signals
fig.add_trace(go.Scatter(x=data[data['Position'] == 2].index,
                         y=data['Fast_MA'][data['Position'] == 2],
                         mode='markers', name='Buy Signal', marker=dict(size=10, color='gold')))

# add sell signals
fig.add_trace(go.Scatter(x=data[data['Position'] == -2].index,
                         y=data['Fast_MA'][data['Position'] == -2],
                         mode='markers', name='Sell Signal', marker=dict(size=10, color='black')))

# plot
fig.update_layout(title='Moving Average Crossover Strategy with Buy/Sell Signals',
                  xaxis_title='Date',
                  yaxis_title='Price',
                  legend_title='Legend',
                  template='plotly_dark')

# Show the plot
fig.show()

```

Bollinger Bands with Relative Strength Index (RSI) indicators

```python

# download stock data
data = yf.download('AAPL', start='2015-01-01', end='2024-01-01')

# dynamic window based on 30-day volatility
data['Volatility'] = data['Close'].rolling(window=30).std()
data['Window'] = (data['Volatility'] / data['Volatility'].mean() * 20).round()
data['Window'] = data['Window'].fillna(20).astype(int)  # fill NaN with 20 and convert to int
data['Window'] = data['Window'].apply(lambda x: max(min(x, 60), 5))  # limit the size between 5 and 60 days

# calculate Bollinger Bands with dynamic window
def bollinger_bands(row, data):
    window = int(row['Window'])
    std = data['Close'].rolling(window=window).std()[row.name]
    mean = data['Close'].rolling(window=window).mean()[row.name]
    return pd.Series([mean + 2*std, mean - 2*std], index=['Upper_Band', 'Lower_Band'])

data[['Upper_Band', 'Lower_Band']] = data.apply(bollinger_bands, axis=1, data=data)

# calculate RSI
delta = data['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
data['RSI'] = 100 - (100 / (1 + rs))

# generate signals with Bollinger Bands and RSI confirmation
conditions = [
    (data['Close'] < data['Lower_Band']) & (data['RSI'] < 30),  # Buy when price is below lower band and RSI is below 30
    (data['Close'] > data['Upper_Band']) & (data['RSI'] > 70)   # Sell when price is above upper band and RSI is above 70
]
choices = [1, -1]
data['Signal'] = np.select(conditions, choices, default=0)

# plot results
fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], name='Close'))
fig.add_trace(go.Scatter(x=data.index, y=data['Upper_Band'], name='Upper Band', line=dict(color='red', dash='dash')))
fig.add_trace(go.Scatter(x=data.index, y=data['Lower_Band'], name='Lower Band', line=dict(color='green', dash='dash')))
fig.add_trace(go.Scatter(x=data.index, y=data['RSI'], name='RSI', yaxis='y2'))

fig.update_layout(title='AAPL Bollinger Bands with RSI and Dynamic Window',
                  xaxis_title='Date',
                  yaxis_title='Price',
                  yaxis2=dict(title='RSI', overlaying='y', side='right'),
                  template='plotly_dark')
fig.show()

```

#Introduction to backtesting:

backtesting is how you compare your trading strategy to see if it would be useful (or essentially more useful than buy & hold).

check out the documentation for [backtesting](https://pypi.org/project/Backtesting/)

```python
# install backtesting package
!pip install backtesting
```

```python
# import dependencies
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
```

```python
# download AAPL stock data
start = "2015-01-01"
end = "2024-01-01"
stock = "AAPL"

data = yf.download(stock, start=start, end=end)
```

```python
# initialize strategy class
class MySMAStrategy(Strategy):
  def init(self):
    price = self.data.Close
    self.ma1 = self.I(SMA, price, 10)
    self.ma2 = self.I(SMA, price, 20)

# define your trading strategy here, using simple crossover trading for example
  def next(self):
    if crossover(self.ma1, self.ma2):
      self.buy()
    elif crossover(self.ma2, self.ma1):
      self.sell()
```

```python
backtest = Backtest(data, MySMAStrategy, commission=.002, exclusive_orders=True)
stats=backtest.run()
```

```python
# evaluate your strategy
print(stats)
```

```python
# plot your strategy
backtest.plot()
```
